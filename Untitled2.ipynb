```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# To ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset
file_path = 'globalterrorismdb_0718dist.csv'
data = pd.read_csv(file_path, encoding='latin1')
```

```python
# Display the first few rows of the dataset
data.head()

# Display basic information about the dataset
data.info()

# Check for missing values
data.isnull().sum()
```

```python
# Convert 'iyear' to datetime format for plotting
data['iyear'] = pd.to_datetime(data['iyear'], format='%Y')

# Plot the number of attacks per year
plt.figure(figsize=(10, 6))
sns.countplot(x='iyear', data=data, palette='viridis')
plt.title('Number of Terrorist Attacks Each Year')
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Plot the number of attacks by country
plt.figure(figsize=(15, 8))
country_counts = data['country_txt'].value_counts().head(20)
sns.barplot(x=country_counts.index, y=country_counts.values, palette='viridis')
plt.title('Top 20 Countries with the Most Terrorist Attacks')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Plot the number of attacks by attack type
plt.figure(figsize=(15, 8))
attack_counts = data['attacktype1_txt'].value_counts()
sns.barplot(x=attack_counts.index, y=attack_counts.values, palette='viridis')
plt.title('Types of Terrorist Attacks')
plt.xticks(rotation=90)
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.show()

# Plot the number of attacks by target type
plt.figure(figsize=(15, 8))
target_counts = data['targtype1_txt'].value_counts().head(20)
sns.barplot(x=target_counts.index, y=target_counts.values, palette='viridis')
plt.title('Types of Targets in Terrorist Attacks')
plt.xticks(rotation=90)
plt.xlabel('Target Type')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Plot the geographical distribution of attacks
plt.figure(figsize=(15, 10))
world_map = sns.scatterplot(x='longitude', y='latitude', hue='region_txt', data=data, palette='viridis', legend=False)
plt.title('Geographical Distribution of Terrorist Attacks')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
